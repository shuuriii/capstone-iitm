# Phase 6 Monitoring, Drift Detection, and Governance

## Business Goal

Ensure long-term reliability, safety, and compliance of the hospital AI system. Phase 6 introduces automated validation for incoming data, drift tracking for features and predictions, and audit logging for traceability.

## Deliverable 1: Monitoring Script

Monitoring script:

```text
scripts/monitor_phase6.py
```

The script performs:

- Missing-value checks for critical operational and billing fields.
- Numeric range checks for age, billing, length of stay, utilization, calendar, and ratio fields.
- Unseen-category checks for department, visit type, insurance provider, city, risk score, and claim status.
- Numeric feature drift using population stability index.
- Categorical feature drift using maximum category share difference.
- Prediction audit-log summary using `logs/predictions.jsonl`.

Run command:

```bash
python3 scripts/monitor_phase6.py
```

For a future production batch:

```bash
python3 scripts/monitor_phase6.py --current-data path/to/new_batch.csv
```

## Governance and Compliance Document

Governance document:

```text
docs/phase6/governance_compliance.md
```

Implemented governance controls:

- Data validation before trusting model inputs.
- Drift tracking before assuming future data matches training data.
- Prediction audit logging with model versions and input hashes.
- Human-in-the-loop recommendation for operational and financial decisions.
- Retraining triggers based on drift, payer behavior changes, and critical-class recall degradation.

Limitations and assumptions:

- PSI and category-share drift are monitoring indicators, not causal explanations.
- Drift thresholds should be tuned with real production history.
- Logs store input hashes rather than raw features to reduce sensitive-data exposure.
- Current models remain decision-support tools and require human review.

## Retraining Strategy

Recommended retraining approach:

- Review monitoring outputs monthly.
- Retrain when a feature has high drift or multiple features have moderate drift.
- Retrain after major payer contract changes or department workflow changes.
- Rerun Phase 4 evaluation before promoting new model artifacts.
- Update the model card and deployment documentation after each promoted retraining cycle.
